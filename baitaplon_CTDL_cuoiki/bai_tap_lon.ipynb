```python
from google.colab import drive
drive.mount('/content/drive')
# choose == 1
def search(dic,my_file):
    word_find = input('Hãy nhập từ vựng bạn cần tra từ điển: ')
    dic = read_dic(dic, my_file)
    if word_find in dic:
        for key, value in dic.items():
          if key == word_find:
            for v in value:
              print(f"- {v}")
            break
    else:
        print("Không tìm thấy từ bạn muốn tra cứu!")

def read_dic(dic, my_file):
    with open(my_file, 'r') as file:
        for line in file:
            if ':' in line:
                key, value = line.split(':', 1)
                key = key.strip()
                value = value.strip()
                if key in dic:
                    if not isinstance(dic[key], list):
                        dic[key] = [dic[key]]
                    dic[key].append(value)
                else:
                    dic[key] = [value]
    return dic

def add_word(my_file, new_word, new_meaning):
    with open(my_file, 'a') as file:
      file.write(f"{new_word}: {new_meaning}\n")
def sort_dic(dic):
    sorted_keys = sorted(dic.keys())
    sorted_dict = {key: dic[key] for key in sorted_keys}
    return sorted_dict

# choose == 3
def delete(dic,my_file):
    dic = read_dic(dic, my_file)
    word = input("Nhập vào từ cần loại bỏ khỏi từ điển: ")
    if word in dic:
        del dic[word]
        print(f"{word} - Đã được loại bỏ!")
        save_dic_to_file(dic, my_file)
    else:
        print(f"{word} không có trong từ điển.")
    return dic


# choose == 4
def save_dic_to_file(dic, my_file):
  with open(my_file, 'w') as file:
      for key, values in dic.items():
        for value in values:
          file.write(f"{key}: {value}\n")

# choose == 5
def load_dic_to_file(dic, my_file):
  with open(my_file, 'r') as file:
      for line in file:
          key, value = line.split(':')
          dic[key] = value
      return dic

def menu():
    print("\n\n")
    string = "DANH MỤC TỪ ĐIỂN"
    print(f"{string.center(40)}\n")
    print("1-Tra cứu từ vựng.\n")
    print("2-Thêm từ vựng.\n")
    print("3-Loại bỏ từ vựng.\n")
    print("4-Lưu từ điển vào tập tin.\n")
    print("5-Nạp từ điển từ các tập tin.\n")
    print("6-Kết thúc chương trình.\n")
    print("\n\n")

def main():
  dic = {}
  my_file = 'N21DCDT096_mang.txt'
  while True:
    choise = int(input("\nNhập lựa chọn mà bạn muốn: "))

    if choise == 1:              #Tra cứu từ điển
      search(dic,my_file)

    elif choise == 2:            #Thêm một mục từ vào từ điển
      dic = read_dic(dic,my_file)
      new_word = input("Nhập từ bạn muốn thêm vào từ điển: ")
      new_meaning = input("Nghĩa của từ: ")
      add_word(my_file, new_word, new_meaning)

      update_dic = read_dic({},my_file)
      sorted_dic = sort_dic(update_dic)
      print("\nTừ điển sau khi thêm và sắp xếp:")
      for key, values in sorted_dic.items():
        for value in values:
           print(f"{key}: {value}")

    elif choise == 3:            #Loại mục từ ra khỏi từ điển
      delete(dic,my_file)

    elif choise == 4:            #Lưu từ điển vào các tập tin
      dic = read_dic({}, my_file)
      sort_dictionary = sort_dic(dic)
      filename1 = input("Nhập tên tập tin để lưu dictionary hoặc tên tập tin mới được tạo tự động: ")
      save_dic_to_file(sort_dictionary,filename1 )
      print(f"Đã lưu vào trong tập tin {filename1}.")

    elif choise == 5:             #Nạp từ điển từ các tập tin
      dictionary = dic
      filename2 = input("Hãy nhập tên file để nạp vào từ điển: ")
      try:
        load_dic_to_file(dictionary,filename2)
        print(f"Đã nạp từ điển từ tập tin {filename2}. ")
        print(f"Từ điển hiện tại là: {dictionary}.")
      except:
        print(f"Tập tin {filename2} không tồn tại trong driver!")

    elif choise == 6:              #Kết thúc chương trình
      print(f"Đã đóng!\nCám ơn bạn đã sử dụng!.")
      break

    else:
      print("Lựa chọn không đúng, vui lòng nhập lại!")
#-------------------------------------------------------------------------------
menu()
main()


#hoàn chỉnh 2:19 15/5
```
